```python
import matplotlib.pyplot as plt

labels = ['Apple', 'Lumia', 'Nokia', 'Samsung']
sizes = [45, 10, 20, 30]
colors =['red', 'blue', 'yellow', 'green']

figure = plt.figure(figsize=(5, 5))
axes = figure.add_subplot(1,1,1)
axes.pie(sizes, colors= colors, labels=labels, textprops={'fontweight': 'bold'}, autopct='%1.1f%%')
plt.show()

```

```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors =['yellow', 'green', 'red', 'blue']
explode = [0.2, 0, 0, 0]  

figure = plt.figure(figsize=(5, 5))
axes = figure.add_subplot(1, 1, 1)
axes.pie(values, colors=colors, labels=labels, shadow= True, explode=explode, textprops={'fontweight': 'bold'}, autopct='%1.1f%%', startangle=180)

plt.show()
```
