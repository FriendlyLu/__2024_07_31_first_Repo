```python
import random
with open('names.txt',encoding='utf-8') as file:
    names_str = file.read()
    display(names_str)
names_list = names_str.split(sep='\n')
names = random.choices(names_list,k=50)
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
df1 = pd.DataFrame(scores,
                   columns=['國文','英文','數學','地理','探索'],
                   index=names)
df1
```

```python
#to_numpy
df1.to_numpy()
```

```python
#to csv
df1.to_csv('student.csv')
```

```python
#to excel
df1.to_excel('student.xlsx', sheet_name='3年5班')
```
